```python
def load_pickle_in_chunks(file_path, chunk_size):
    chunks = []
    with open(file_path, 'rb') as file:
        while True:
            try:
                chunk = []
                for _ in range(chunk_size):
                    chunk.append(pickle.load(file))
                chunks.append(chunk)
            except EOFError:
                break
    return chunks

# Example usage
file_path = os.path.join(log_dir, "676b44e51d6c9145937ba8176de28971.results.test_censored_attribution_0")
chunk_size = 10  # Adjust the chunk size as needed
chunks = load_pickle_in_chunks(file_path, chunk_size)
```
